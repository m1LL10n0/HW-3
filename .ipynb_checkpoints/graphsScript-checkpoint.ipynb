```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('printOut.csv')
```

```python
cacheline = data[(data.test == 'cacheline32') | (data.test == 'cacheline64') | (data.test == 'cacheline128')]

plt.style.use('ggplot')

cache=cacheline.pivot(index='spec', columns='test', values='area')
cache=cache.reindex(columns=['cacheline32','cacheline64','cacheline128'])

cache.plot(kind='bar', figsize=(8,5), width=0.7)

plt.grid(True)
plt.title('Area - Cacheline size')
plt.xlabel('')
plt.ylabel('mm^2')
plt.xticks(fontsize=12,rotation=0)

plt.show()
```

```python
cacheline = data[(data.test == 'l1d32') | (data.test == 'l1d64') | (data.test == 'l1d128')]

plt.style.use('ggplot')

cache=cacheline.pivot(index='spec', columns='test', values='area')
cache=cache.reindex(columns=['l1d32','l1d64','l1d128'])

cache.plot(kind='bar', figsize=(8,5), width=0.7)

plt.grid(True)
plt.title('Area - L1d size')
plt.xlabel('')
plt.ylabel('mm^2')
plt.xticks(fontsize=12,rotation=0)

plt.show()
```

```python
conditions=[(data['test'] == 'cacheline32') | (data['test'] == 'l1d32') | (data['test'] == 'l1dassoc1') | (data['test'] == 'l1i32') | (data['test'] == 'l1iassoc1') | (data['test'] == 'l2512') | (data['test'] == 'l2assoc1'),
            (data['test'] == 'cacheline64') | (data['test'] == 'l1d64') | (data['test'] == 'l1dassoc2') | (data['test'] == 'l1i64') | (data['test'] == 'l1iassoc2') | (data['test'] == 'l21024') | (data['test'] == 'l2assoc2'),
            (data['test'] == 'cacheline128') | (data['test'] == 'l1d128') | (data['test'] == 'l1dassoc4') | (data['test'] == 'l1i128') | (data['test'] == 'l1iassoc4') | (data['test'] == 'l24096') | (data['test'] == 'l2assoc4')]
choices=['1','2','3']
data['testVar']= np.select(conditions,choices)

conditions2=[(data['test'] == 'cacheline32') | (data['test'] == 'cacheline64') | (data['test'] == 'cacheline128'),
             (data['test'] == 'l1d32') | (data['test'] == 'l1d64') | (data['test'] == 'l1d128'),
             (data['test'] == 'l1dassoc1') | (data['test'] == 'l1dassoc2') | (data['test'] == 'l1dassoc4'),
             (data['test'] == 'l1i32') | (data['test'] == 'l1i64') | (data['test'] == 'l1i128'),
             (data['test'] == 'l1iassoc1') | (data['test'] == 'l1iassoc2') | (data['test'] == 'l1iassoc4'),
             (data['test'] == 'l2512') | (data['test'] == 'l21024') | (data['test'] == 'l24096'),
             (data['test'] == 'l2assoc1') | (data['test'] == 'l2assoc2') | (data['test'] == 'l2assoc4')]

choices2=['cacheline','l1d','l1dassoc','l1i','l1iassoc','l2','l2assoc']
data['testName']=np.select(conditions2,choices2)
```

```python
data1spec = data[data.spec=='specbzip.txt']

data1spec = data1spec.pivot(index='testName',
                            columns='testVar', 
                            values='area')
data1spec.plot(kind='bar', figsize=(8,5), width=0.7)

plt.grid(True)
plt.title('Area variations')
plt.xlabel('')
plt.ylabel('mm^2')
plt.xticks(fontsize=12,rotation=0)

plt.show()
```
